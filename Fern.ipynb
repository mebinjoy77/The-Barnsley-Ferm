```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```
def function(x,y):
  return (0.,0.16*y)
def function1(x,y):
  return (0.85*x+0.04*y,-0.04*x+0.85*y + 1.60)
def function2(x,y):
  return (0.20*x +	-0.26*y,	0.23*x	+ 0.22*y +	1.60)
def function3(x,y):
  return(-0.15*x +	0.28*y,	0.26*x +	0.24*y + 0.44)
functions = [function,function1,function2,function3]
```

```
width = 500
height = 500
points = 100000
x,y = 0,0
photo = np.zeros((width,height))
for i in range(points):
  function = np.random.choice(functions,p=[0.01,0.85,0.07,0.07])
  x,y = function(x,y)
  ix,iy = int(width/2+ x*width/10),int(y*height/12)
  photo[iy,ix] = 1
plt.imshow(photo[::-1,:],cmap=cm.Blues)
plt.show()
```
